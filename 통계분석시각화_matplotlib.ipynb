#plot 함수

```python
import numpy as np
data = np.arange(10) #y축 데이터
data
```

```python
plt.plot(data) #도면에 추가, x축 데이터는 0부터 순차적으로 자동 부여
plt.show() #시각화
```

```python
#y값만 인자로 전달, x축 데이트는 0부터 순차적으로 자동 부여
plt.plot([1,4,7,9,12,15])
plt.show()
```

```python
#제목 부여
plt.plot([1,4,7,9,12,15]) #도면에 추가
plt.title("plt.plot([1,4,7,9,12,15])") #제목 달기
plt.show() #화면에 시각화
```

```python
#x, y전달
plt.plot([1,2,3,4,5,7], [1,2,9,7,12,15]) #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7][1,2.9.7,12,15])") #제목 달기
plt.show() #화면에 시각화
```

```python
#마커 사용: 'o' 원마커
plt.plot([1,2,3,4,5,7], [1,2,9,7,12,15], 'o') #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7][1,2.9.7,12,15]), 'o'") #제목 달기
plt.show() #화면에 시각화
```

```python
#마커 사용: 'v' 
plt.plot([1,2,3,4,5,7], [1,2,9,7,12,15], 'v') #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7][1,2.9.7,12,15]), 'v'") #제목 달기
plt.show() #화면에 시각화
```

```python
#마커+라인 사용: 'o-'
plt.plot([1,2,3,4,5,7], [1,2,9,7,12,15], 'o-') #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7][1,2.9.7,12,15]), 'o-'") #제목 달기
plt.show() #화면에 시각화
```

```python
#마커+라인 사용: 'o:'
plt.plot([1,2,3,4,5,7], [1,2,9,7,12,15], 'o:') #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7][1,2.9.7,12,15]), 'o:'") #제목 달기
plt.show() #화면에 시각화
```

```python
#마커+라인+색상 사용: 'o:c'
plt.plot([1,2,3,4,5,7], [1,2,9,7,12,15], 'o:c') #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7][1,2.9.7,12,15]), 'o:c'") #제목 달기
plt.show() #화면에 시각화
```

```python
#kwargs의 linewidth 라인 두께
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:', color='#FF00FF', linewidth=5) #도면에 추가
plt.title("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'o:', color='#FF00FF', linewidth=5)") #제목 달기
plt.show() #화면에 시각화
```

```python
#데이터를 두 개 이상 전달하기
plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo', [1,2,3,4,5,7],[8,9,10,15,10,11], 'r^') 
plt.title("plt.plot([1,2,3,4,5,7],[1,2,9,7,12,15],'bo', [1,2,3,4,5,7],[8,9,10,15,10,11], 'r^')")
plt.show() #화면에 시각화  --> 가독성이 좋지 못함
```

```python
#데이터를 두 개 이상 전달하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo')
plt.plot([1,2,3,4,5],[95,81,87,78,94], 'r^')
plt.title("scores") #제목 달기
plt.show() #화면에 시각화
```

```python
#축 범위 지정하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo')
plt.plot([1,2,3,4,5],[95,81,87,78,94], 'r.')
plt.axis([0,6,0,100])
plt.title("scores") #제목 달기
plt.show() #화면에 시각화
```

```python
#축 이름 지정하기
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo')
plt.plot([1,2,3,4,5],[95,81,87,78,94], 'r.')
plt.axis([0,6,0,100])
plt.xlabel('number')
plt.ylabel('score')
plt.title("scores") #제목 달기
plt.show() #화면에 시각화
```

```python
#label 표시
plt.plot([1,2,3,4,5],[90,80,85,77,97],'bo', label='class 1')
plt.plot([1,2,3,4,5],[95,81,87,78,94], 'r.', label='class 2')
plt.axis([0,6,0,100])
plt.xlabel('number')
plt.ylabel('score')
plt.title("scores") #제목 달기
plt.legend() #label을 표시하세요 (범례)
plt.show() #화면에 시각화
```

```python
# sub 도표 사용하기
plt.subplot(221) #2x2 도표 중에 1
plt.title("221")
plt.plot([1,2,3,4], 'ro-')

plt.subplot(222) #2x2 도표 중에 2
plt.title("222")
plt.plot([4,3,2,1], 'b.:')

plt.subplot(223) #2x2 도표 중에 3
plt.title("223")
plt.plot([1,3,2,4],'c^-.')

plt.subplot(224) ##2x2 도표 중에 4
plt.title("224")
plt.plot([2,4,1,3],'k.--')

plt.show()
```

```python
# y축에 표현할 값을 지정 + 디폴트 색상 확인
ys_str = []
for i in range(1,31): # 1~30
  plt.plot([i,i])
  if i % 3 == 0:
    ys_str.append(str(i))
  else:
    ys_str.append(" ")
plt.title("default color, set yticks")
plt.yticks(range(1,31),ys_str)
plt.show() # 10개까지는 디폴트로 다른 색상이 지정되지만 11개 이후부터는 개발자가 직접 해야 함
```

```python
# 배경색
plt.plot([1,1], 'ro-')
ax = plt.gca()
ax.set_facecolor('violet')
plt.title("set background color")
plt.show()
```

```python
names = ['a','b','c']
scores = [80,90,95]
plt.bar(names,scores) #x,y
plt.xlabel('name')
plt.ylabel('score')
plt.title('name=score bar')
plt.show()

plt.plot(names,scores) #x,y
plt.xlabel('name')
plt.ylabel('score')
plt.title('name=score bar')
plt.show()

plt.barh(names,scores) #x,y
plt.xlabel('name')
plt.ylabel('score')
plt.title('name=score bar')
plt.show()
```

#파이그래프

```python
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
plt.pie(incomes, labels = locales)
plt.title("지역별 수입")
plt.show()
```

```python
#돌출시키기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs=[0,0.9,0,0,0] #돌출시킬 수치
plt.pie(incomes, labels = locales, explode=exs)
plt.title("지역별 수입 - 돌출시키기")
plt.show()
```

```python
#퍼센트 표시하기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs=[0,0.1,0,0,0] #돌출시킬 수치
plt.pie(incomes, labels = locales, explode=exs, autopct='%.2f%%')
plt.title("지역별 수입 - 퍼센트 표시하기")
plt.show()
```

```python
#퍼센트 표시할 위치 정하기
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs=[0,0.1,0,0,0]
plt.pie(incomes, labels = locales, explode=exs, autopct='%.2f%%', pctdistance=0.8)
plt.title("지역별 수입 - 퍼센트 표시할 위치 정하기")
plt.show()
```

```python
#그림자 효과
locales = ['서울','강원','경북','충남','기타']
incomes = [2300,1200,1800,800,400]
exs=[0,0.1,0,0,0]
plt.pie(incomes, labels = locales, explode=exs, autopct='%.2f%%', pctdistance=0.8, shadow = True)
plt.title("지역별 수입 - 그림자 효과")
plt.show()
```

```python
#label 위치 지정
locales = ['서울','강원','경북','충남','기타'] #서울 기준점 3시
incomes = [2300,1200,1800,800,400]
exs=[0,0.1,0,0,0]
plt.pie(incomes, labels = locales, explode=exs, autopct='%.2f%%', pctdistance=0.8, labeldistance=1.15)
plt.title("지역별 수입 - label 표시할 위치 지정하기")
plt.show()
```

```python
#시계 방향으로 각도 계산하기
locales = ['서울','강원','경북','충남','기타'] #startangle default: x축 기준 0도(3시)
incomes = [2300,1200,1800,800,400]
exs=[0,0.1,0,0,0]
plt.pie(incomes, labels = locales, explode=exs, autopct='%.2f%%', startangle=90, counterclock=False)
plt.title("지역별 수입 - 시계 방향으로 각도 계산하기")
plt.show()
```
