```python
height = [165, 177, 160, 180, 185, 155, 172]
weight = [62,67,55,74,90,43,64]

plt.plot(height,weight,'ro')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid(True)
plt.title("Height-Weight")
plt.show()
```

```python
plt.scatter(height,weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid(True)
plt.title("Height-Weight")
plt.show()
```

```python
plt.scatter(height,weight, s=500) # 전부 scale 500으로 지정 (다 다르게도 가능)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid(True)
plt.title("Height-Weight")
plt.show()
```

```python
size = 100 * np.arange(1,8)
colors = ['r','g','b','c','m','k','y']
plt.scatter(height,weight,s=size, c=colors)
plt.show()
```

```python
import numpy as np
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

# 인구 밀도(명/km^2): 2017년 통계청 자료
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

# 위도(latitude)와 경도(longitude)
lat = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

size = np.array(pop_den) * 0.2  # 마커의 크기 지정 
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커의 색 지정 

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5) #alpha : 투명도
plt.xlabel('경도(lonitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

# 지역명 출력 
for x, y, name in zip(lon, lat, city) : 
    plt.text(x, y, name) 

plt.show()
```

#막대그래프

```python
ids = ['m01','m02','m03','m04'] #회원 아이디
be_ex = [27,35,40,33] #운동시작전
af_ex = [30,38,42,37] #운동한달후
n = len(ids) #회원 4명
index = np.arange(n) #Numpy를 이용해 배열 생성 [0,1,2,3]
plt.bar(index,be_ex) # bar(x,y)에서 x=index, height=before_ex로 지정
plt.grid(True)
plt.show()
```

```python
plt.bar(index,be_ex,tick_label=ids) #x축의 tick_label은 숫자로 되어있음 -> 회원id로 지정
plt.show()
```

```python
plt.bar(ids,be_ex) 
plt.show()
```

```python
colors =['r','g','b','m']
plt.bar(ids,be_ex,color=colors) # 색상 변경
plt.show()
```

```python
colors =['r','g','b','m']
plt.bar(ids,be_ex,color=colors, width=0.6) #막대 그래프 폭 변경 (barh는 width 변경 불가)
plt.show()
```

```python
colors =['r','g','b','m']
plt.barh(ids,be_ex,color=colors, height=0.6) #barh --> height으로 폭 변경
plt.show()
```

```python
bw = 0.4 # 막대그래프 폭 (두개가 들어갈 수 있게 0.4로)
plt.bar(index,be_ex,color='c',align='edge',width=bw,label='before')
plt.bar(index+bw,af_ex,color='m',align='edge',width=bw,label='after')
plt.xticks(index+bw, ids)
plt.title("운동 전 후 윗몸 일으키기 횟수")
plt.legend()
plt.show()
```
