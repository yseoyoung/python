```python
# Mr, Ms 가 각각 몇 명인지 파악 시도해볼 수 있음

names_values=['None','Mr','Mrs','Miss','Master','Dr']
ndata = data['Name']
for ne in ndata:
  f,s = ne.split(',') # 이름을 콤마를 기준으로 분리시킴 (요소 두개 나옴)
  ses = s.split('.') # 분리된 이름 중 뒷 부분을 .을 기준으로 자르기 (.이 없는 것도 있으니 리스트 형태로 저장한 후, 맨 앞 요소만)
  s0 = ses[0] # 앞 부분(명칭=인덱스 0)만 출력
  s0 = s0.replace(' ','') #공백 제거
  #print(s0)

  check=False #names_values 에 해당하는 값이 없으면 false
  for i,ne2 in enumerate(names_values): # enumertae는 인덱스 값, 원소 두 개를 동시에 설정할 수 있음
    if ne2 == s0: #s0의 요소가 names_values의 요소와 같은 것이 있으면
      print(i) #해당하는 인덱스 출력
      check = True #names_values 에 해당하는 값이 있으면 true
      break
  if check == False:
    print(0)

# 이름데이터를 우리가 알고 있는 유의미한 범주형 데이터로 변경하였다 = <<전처리 과정>>
# 이 명칭은 생존 여부를 분석하는 데에 중요한 단서가 된다.
```
